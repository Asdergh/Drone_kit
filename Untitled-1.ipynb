Первая модель нейронной сетию. Классический перцептрон

```python
class Perceptron():
    def __init__(self, eta, epochs) -> None:
        self.eta = eta
        self.epochs = epochs
    
    def fit(self, X, Y):
        self.W = np.zeros(1 + X.shape[1])
        self.errors_ = []

        for _ in range(self.epochs):
            errors = 0
            for (xi, true_label) in zip(X, Y):

                update = self.eta * (true_label - self.prediction(xi))
                self.W[1:] += update * xi
                self.W[0] += update
                errors += int(update != 0.0)
                
            self.errors_.append(errors)
        return self
    def net_input(self, X):
        return np.dot(X, self.W[1:]) + self.W[0]

    def prediction(self, X):
        return np.where(self.net_input(X) >= 0.0, 1, -1)
```

создадим функцию для отрисовки полученного из n - классов решений нащим перцептроном

```python
def generate_desition_plot(X, Y, classifier,  frequency=0.02):
    markers = ["s", "x", "^", "v", "o"]
    colors = ["blue", "red", "lightgreen", "gray", "cyan"]
    cmap = mcolors.ListedColormap(colors[:len(np.unique(Y))] )

    x1_min = X[:, 0].min() - 1
    x1_max = X[:, 0].max() + 1
    x2_min = X[:, 1].min() - 1
    x2_max = X[:, 1].max() + 1

    xx1_grid, xx2_grid = np.meshgrid(np.arange(x1_min, x1_max, frequency),
    np.arange(x2_min, x2_max, frequency))

    Z = classifier.prediction(np.array([xx1_grid.ravel(), xx2_grid.ravel()]).T)
    Z = Z.reshape(xx1_grid.shape)

    plt.contourf(xx1_grid, xx2_grid, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1_grid.min(), xx1_grid.max())
    plt.ylim(xx2_grid.min(), xx2_grid.max())

    for (index, cls) in enumerate(np.unique(Y)):
        plt.scatter(x=X[Y==cls, 0], y=X[Y==cls, 1], alpha=0.8, c=cmap(index), marker=markers[index], label=cls)
    
```

```python
generate_desition_plot(X, y, classifier=pn)
plt.xlabel("длинна чашелистиника")
plt.ylabel("длинна лепестка")
plt.legend(loc="upper left")
plt.show()
```

теперь опробуем новую модель машинного обучения под названием линейный перцептрон
данный вид перцептрона отличается от самого перцептрона тем что у него есть дополниельный линейный нейрон
еще данный тип обучения называется линейной регрессией

```python
class AdalineGd():

    def __init__(self, eta, epochs) -> None:
        self.eta = eta
        self.epochs = epochs
    
    def fit(self, X, Y):

        self.W_ = np.zeros(1 + X.shape[1])
        self.cost_ = []
        
        for _ in range(self.epochs):
            output = self.net_input(X)
            errors = (Y - output)
            self.W_[1:] += self.eta * X.T.dot(errors)
            self.W_[0] += self.eta * errors.sum()
            cost = (errors ** 2).sum() / 2.0
            self.cost_.append(cost)
        return self
    
    def net_input(self, X):
        return np.dot(X, self.W_[1:]) + self.W_[0]
    
    def activation(self, X):
        return self.net_input()
    
    def prediction(self, X):
        return np.where(self.net_input(X) >= 0.0, 1, -1)
```

```python
class Last_Neuron():
    def __init__(self, eta, epochs) -> None:
        self.eta = eta
        self.epochs = epochs
    
    def fit(self, X, Y):

        self.W_ = np.zeros(1 + X.shape[1])
        self.cost_ = []

        for _ in range(self.epochs):

            cost = 0
            output = self.net_input(X)
            errors = (Y - output)
            
            self.W_[1:] += self.eta * X.T.dot(errors)
            self.W_[0] += self.eta * errors.sum()

            cost += (errors ** 2) / 2.0
            self.cost_.append(cost)
        return self
    
    def net_input(self, X):
        return np.dot(X, self.W_[1:]) + self.W_[0]

    def prediction(self, X):
        return np.where(self.net_input(X) >= 0.0, -1, 1)
```

```python
class AdalineSGD(object):

    def __init__(self, eta = 0.01, epochs=10, shuffle=True, random_seet=None) -> None:
        self.eta = eta
        self.epochs = epochs
        self.shuffle = shuffle
        self.random_state = random_seet
        if self.random_state:
            np.random.seed(random_seet)
    
    def fit(self, X, y):

        self._initialize_weights(X.shape[1])
        self.cost_ = []

        for i in range(self.epochs):
            if self.shuffle:
                X, y = self._shuffle(X, y)
            cost = []
            for (xi, true_label) in zip(X, y):
                cost.append(self._update_weight(xi, true_label))
            avg_cost = sum(cost) / len(y)
            self.cost_.append(avg_cost)
            
        return self
    
    def partial_fit(self, X, y):

        if not self.W_initialised:
            self._initialize_weights(X.shape[1])
        if y.ravel().shape[1] > 1:
            for (xi, true_label) in zip(X, y):
                self._update_weight(xi, true_label)
        else:
            self._update_weight(X, y)
        
        return self
            

    def _initialize_weights(self, m):
        self.W_ = np.zeros(1 + m)
        self.W_initialised = True
    


    def _shuffle(self, X, y):
        rand = np.random.permutation(len(y))
        return X[rand], y[rand]
    
    def net_input(self, X):
        return np.dot(X, self.W_[1:]) + self.W_[0]
    
    def _update_weight(self, xi, true_label):
        output = self.net_input(xi)
        error = (true_label - output)

        self.W_[1:] += self.eta * xi.dot(error)
        self.W_[0] += self.eta * error

        cost = 0.5 * error ** 2

        return cost

    
    def activation(self, X):
        return self.net_input(X)
    
    def prediction(self, X):
        return np.where(self.activation(X) >= 0.0, 1, -1)
```

```python
Ada = AdalineSGD(eta=0.01, epochs=20, random_seet=1)
Ada.fit(X_std, y)
generate_desition_plot(X_std, y, classifier=Ada)
plt.legend(loc="upper left")
plt.title("ADALINESGD")
plt.show()
plt.plot(range(1, len(Ada.cost_) + 1), Ada.cost_, marker="o")
plt.show()
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import sklearn.linear_model as linear_model


iris = datasets.load_iris()
X = iris.data[:, [2, 3]]
Y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)

sc = StandardScaler().fit(X_train)

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

ppn = linear_model.Perceptron(max_iter=40, eta0=0.001,  random_state=0)
ppn.fit(X_train_std, y_train)

y_pred = ppn.predict(X_test_std)
print(f"число ошибочных классификаций: {(y_test != y_pred).sum()}")


```
